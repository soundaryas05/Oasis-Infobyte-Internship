
# Exploratory Data Analysis (EDA) on Retail Sales Data

## Project Overview

In this project, we perform **Exploratory Data Analysis (EDA)** on a **Retail Sales dataset**. The goal is to uncover patterns, trends, and insights that can help the retail business optimize its operations and make data-driven decisions.

---

## Dataset

The project utilizes two datasets that contain information about retail sales, customer demographics, and product details. 

- **Dataset 1**: https://www.kaggle.com/datasets?search=retail+sales



## Key Concepts and Challenges

1. **Data Loading and Cleaning**: 
   - Import and load the retail sales dataset.
   - Clean the dataset by handling missing values, removing duplicates, and formatting columns.
   
2. **Descriptive Statistics**: 
   - Calculate key statistics such as **mean**, **median**, **mode**, **standard deviation** to understand the data better.
   
3. **Time Series Analysis**: 
   - Explore sales trends over time using techniques like resampling and rolling averages.
   
4. **Customer and Product Analysis**: 
   - Investigate customer demographics and purchasing behavior, including **age group** and **gender** analysis.
   - Analyze **Product Category** and total sales to understand product performance.
   
5. **Data Visualization**: 
   - Visualize insights using various plots (e.g., **bar charts**, **line plots**, **heatmaps**, etc.) to convey findings effectively.
   
6. **Recommendations**: 
   - Provide actionable recommendations to the retail business based on the findings from the EDA.

---

## Learning Objectives

By completing this project, you will:

- Gain hands-on experience in **data cleaning** and **exploratory data analysis (EDA)**.
- Learn how to calculate and interpret **descriptive statistics** and how to perform **time series analysis**.
- Develop skills in using **data visualization tools** such as **Seaborn**, **Matplotlib**, and **Plotly** for communicating insights.
- Improve your ability to make **data-driven recommendations** for business decision-making.

---

## Steps Involved

1. **Data Preprocessing**:
   - Load the data.
   - Handle missing values, duplicates, and format columns correctly.
   
2. **Univariate Analysis**:
   - Explore the distribution of important features like **Total Amount** and **Product Category**.
   
3. **Bivariate Analysis**:
   - Investigate relationships between different features like **Product Category vs. Total Amount**, **Quantity vs. Price per Unit**, etc.
   
4. **Multivariate Analysis**:
   - Perform analysis across multiple features (e.g., **Date, Total Amount, and Quantity**) to uncover monthly sales trends and patterns.
   
5. **Visualizations**:
   - Create various plots to visualize the insights and trends identified during the analysis.
