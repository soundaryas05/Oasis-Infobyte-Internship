# Natural Language Processing

**Data Preprocessing**

We employ Natural Language Processing (NLP) techniques for data preprocessing. NLP is a branch of computer science and artificial intelligence that focuses on enabling computers to understand, interpret, and generate human language. It involves the development of algorithms and models that can analyze and process vast amounts of textual data, helping machines to perform tasks such as language understanding, translation, sentiment analysis, and more.

1. **Tokenization**:  
Tokenization is the process of splitting text into smaller, manageable units such as words, sentences, or phrases to facilitate further analysis.


2. **Removing Stopwords**:  
Removing stopwords involves eliminating common words like "the", "is", and "and" that do not contribute meaningful information for analysis.


3. **Lemmatization / Stemming**:  
Lemmatization and stemming reduce words to their base or root form, with lemmatization considering context and grammar, while stemming simply removes prefixes or suffixes.


4. **Vectorization**:  
Vectorization converts text into numerical representations, such as through **TF-IDF** or word embeddings, allowing machine learning models to analyze semantic relationships in the data.

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download necessary NLTK resources if not already downloaded
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Load the stopwords only once before the loop
stop_words = set(stopwords.words('english'))

text_list = []

# Loop through the reviews
for i in reviews_df['Translated_Review']:
    # Check if the entry is a string
    if isinstance(i, str):
        # Remove non-alphabetic characters
        text = re.sub("[^a-zA-Z]", " ", i)
        
        # Convert text to lowercase
        text = text.lower()
        
        # Tokenize the text into words
        text = nltk.word_tokenize(text)
        
        # Remove stopwords
        text = [word for word in text if word not in stop_words]
        
        # Lemmatize each word
        text = [lemmatizer.lemmatize(word) for word in text]
        
        # Join the words back into a single string
        text = " ".join(text)
        
        # Append the processed text to the list
        text_list.append(text)
    else:
        # If the entry is not a string, append an empty string or handle as needed
        text_list.append("")

# Check the first few processed texts
print(text_list[:5])  # Print the first 5 processed reviews (or adjust as needed)
```

**Vectorization** involves transforming a collection of text documents into numerical feature vectors. This process converts the text data into a format that can be analyzed or used by machine learning models. In the resulting vectorized representation, each column corresponds to a unique word from the vocabulary (bag of words), and each row represents a document (such as a review). The value in a given row-column cell indicates the frequency of the word in that particular document.

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Example text list (replace with your actual data)
# text_list = ["This is the first review.", "This is the second review."]  # Your text list

# Set the maximum number of features
max_features = 200000

# Initialize CountVectorizer with max_features and stop words
cou_vec = CountVectorizer(max_features=max_features, stop_words="english")

# Perform vectorization (keeping it sparse)
sparse_matrix = cou_vec.fit_transform(text_list)

# Get the feature names (words)
all_words = cou_vec.get_feature_names_out()

# Convert sparse matrix to DataFrame directly
words = pd.DataFrame.sparse.from_spmatrix(sparse_matrix, columns=all_words)

# Display the DataFrame
print(words.head())
```
